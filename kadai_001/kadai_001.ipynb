```python
from openpyxl import Workbook
from openpyxl.styles import Font, Alignment
from datetime import datetime


# ブック作成
wb = Workbook()
ws = wb.active
ws.title = "請求書"


# 列幅調整
ws.column_dimensions["A"].width = 5
ws.column_dimensions["B"].width = 18
ws.column_dimensions["C"].width = 12
ws.column_dimensions["D"].width = 12
ws.column_dimensions["E"].width = 12
ws.column_dimensions["F"].width = 12
ws.column_dimensions["G"].width = 12


# タイトル
ws["B2"] = "請求書"
ws["B2"].font = Font(bold=True, size=16)


# 宛先情報
ws["B4"] = "株式会社ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["B7"] = "担当者名:鈴木一郎 様"

for row in range(4, 8):
    ws[f"B{row}"].font = Font(bold=True, size=12)


# 請求書番号・日付
ws["F4"] = "No."
ws["G4"] = "0001"
ws["F5"] = "日付"
ws["G5"] = datetime.today().strftime("%Y/%m/%d")

for cell in ["F4", "G4", "F5", "G5"]:
    ws[cell].font = Font(bold=True, size=12)


# 明細ヘッダー
headers = ["商品名", "数量", "単価", "金額"]
for col, header in zip(["B", "C", "D", "E"], headers):
    ws[f"{col}10"] = header
    ws[f"{col}10"].font = Font(bold=True, size=12)


# 明細
items = [
    ["商品A", 2, 10000],
    ["商品B", 1, 15000]
]

start_row = 11

for i, item in enumerate(items):
    row = start_row + i
    ws[f"B{row}"] = item[0]
    ws[f"C{row}"] = item[1]
    ws[f"D{row}"] = item[2]
    ws[f"E{row}"] = f"=C{row}*D{row}"


# 小計・消費税・合計
ws["B15"] = "小計"
ws["E15"] = "=SUM(E11:E12)"

ws["B16"] = "消費税"
ws["E16"] = "=E15*0.1"

ws["B17"] = "合計"
ws["E17"] = "=E15+E16"

for row in range(15, 18):
    ws[f"B{row}"].font = Font(bold=True, size=12)
    ws[f"E{row}"].font = Font(bold=True, size=12)


# 右寄せ
for cell in ["C11", "C12", "D11", "D12", "E11", "E12", "E13", "E15", "E16", "E17"]:
    ws[cell].alignment = Alignment(horizontal="right")


# ファイル名
today = datetime.today().strftime("%Y%m%d")
file_name = f"請求書_{today}.xlsx"

wb.save(file_name)

print(f"{file_name} を作成しました")
```
