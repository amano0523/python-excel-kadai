```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill


# Excelファイルを読み込む
df_2022 = pd.read_excel("2022_年間売上表.xlsx", sheet_name="Sheet1")
df_2023 = pd.read_excel("2023_年間売上表.xlsx", sheet_name="Sheet1")


# データを連結
df = pd.concat([df_2022, df_2023], ignore_index=True)


# 商品・売上年ごとにグループ化して金額を合計
summary = df.groupby(
    ["商品", "売上年"],
    as_index=False
)["金額（千円）"].sum()


# Excelファイルに書き込み
summary.to_excel("売上集計表.xlsx", index=False)


# openpyxlで既存Excelファイルを読み込む
workbook = load_workbook("売上集計表.xlsx")
worksheet = workbook.active


# ヘッダー行を薄いグレーに設定
fill = PatternFill(
    fgColor="F2F2F2",
    fill_type="solid"
)

for cell in worksheet[1]:
    cell.fill = fill


# 保存
workbook.save("売上集計表.xlsx")


print("売上集計表.xlsxを作成しました")
print(summary)
```
